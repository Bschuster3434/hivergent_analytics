```python
import pandas
import requests
from requests.compat import urljoin
import datetime
import time
import json
import math
```

```python
#Appears to be a freely available BTC API
#https://btc.com/api-doc#General
```

```python
#Starting Point for analysis
starting_block = 501798 #Dec 31st 2017 start
ending_block = 501960 #Dec 31st 2017 end
get_blocks_by_range = 10
satoshi_to_BTC = 100000000
```

```python
#https://chain.api.btc.com/v3/block/3
url_endpoint = "https://chain.api.btc.com"
```

```python
next_url_call = urljoin(url_endpoint, "v3/block/" + str(starting_block) + "/tx?page=1")
resp = json.loads(requests.get(next_url_call).content)
```

```python
total_iterations = int(math.ceil(resp['data']['total_count']/float(50)))
```

```python
first_resp = resp
first_resp
```

```python
for i in range(2,total_iterations + 1):
    print "Now on iteration: " + str(i)
    next_url_call = urljoin(url_endpoint, "v3/block/" + str(starting_block) + "/tx")
    next_url_call = urljoin(next_url_call, "?page=" + str(i))
    next_resp = json.loads(requests.get(next_url_call).content)
    
    first_resp['data']['list'] +=  next_resp['data']['list']
    
    time.sleep(1)
```
