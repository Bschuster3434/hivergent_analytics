```python
import pandas as pd
import requests
import datetime
import time
import json
import sys
```

```python
#https://min-api.cryptocompare.com/data/histoday?fsym=OMG&tsym=USD&limit=365&aggregate=1&e=CCCAGG

#Target Unit Timestamp
#1514678400

def get_cryptocompare_data(sym):
    '''
    Grabs the day of data being searched for in the symbol.
    Returns the High Value for that Day.
    '''
    url_endpoint = "https://min-api.cryptocompare.com/data/histoday"
    params = {"fsym":sym, \
                "tsym":"USD", \
                "limit" : "365", \
                "aggregate": "1", \
                "e": "CCCAGG"}

    resp = requests.get(url_endpoint, params=params)
    resp_dict = json.loads(resp.content) #Get a string, must eval to a dict
    if resp_dict["Data"] == []:
        return None
    else:
        for i in resp_dict["Data"]:
            if i['time'] == 1514678400:
                return i['high']
    return None
```
