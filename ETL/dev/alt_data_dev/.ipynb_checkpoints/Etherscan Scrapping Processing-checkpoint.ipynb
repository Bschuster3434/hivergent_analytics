```python
import pandas as pd
import requests
import datetime
import time
import json
import sys
import math
```

```python
df = pd.read_csv('123117_Etherscan_Extract_Data.csv')
df = df.fillna('nan')
```

```python
processed_contracts = []
for i in range(len(df)):
    next_address = df.iloc[i]
    next_tx = {}
    
    #Contract Processing
    c = next_address['contract_address']
    c = c.replace('\xc2', ' ')
    c = c.replace('\xa0', ' ')
    c = c.strip().split(' ')
    
    next_tx['smart_contract_hash'] = c[-1]
    
    #Token Processing
    t = next_address['token_name']
    
    if t == 'nan':
        pass
    else:
        token_name = t.split()[-1]
        token_name = token_name.replace('(', '').replace(')','')
        if token_name == "ERR":
            pass
        else:
            next_tx['token_name'] = token_name
            next_tx['etherscan_token_url'] = "https://etherscan.io/token/" + next_tx['smart_contract_hash']
        
    #Contract name
    if next_address['smart_contract_name'] != 'nan':
        next_tx['smart_contract_name'] = next_address['smart_contract_name']
    else:
        try:
            next_tx['smart_contract_name'] = next_tx['token_name'] + "_token_contract"
        except KeyError:
            pass
        
    #Append
    processed_contracts.append(next_tx)
```

```python
pd.DataFrame(processed_contracts).to_csv('123117_etherscan_processed.csv')
```

```python
dec_df = pd.read_csv('123117_etherscan_processed_with_decimal.csv').fillna('nan')
```

```python
#https://min-api.cryptocompare.com/data/histoday?fsym=OMG&tsym=USD&limit=365&aggregate=1&e=CCCAGG

#Target Unit Timestamp
#1514678400

def get_cryptocompare_data(sym):
    '''
    Grabs the day of data being searched for in the symbol.
    Returns the High Value for that Day.
    '''
    url_endpoint = "https://min-api.cryptocompare.com/data/histoday"
    params = {"fsym":sym, \
                "tsym":"USD", \
                "limit" : "365", \
                "aggregate": "1", \
                "e": "CCCAGG"}

    resp = requests.get(url_endpoint, params=params)
    resp_dict = json.loads(resp.content) #Get a string, must eval to a dict
    if resp_dict["Data"] == []:
        return None
    else:
        for i in resp_dict["Data"]:
            if i['time'] == 1514678400:
                return i['high']
    return None
```
